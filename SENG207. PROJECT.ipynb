```python
# Calculate frequencies for each column
frequencies = df.apply(pd.Series.value_counts).fillna(0)

# Create a scatter plot for each column
for column in df.columns:
  mplt.figure()
  mplt.scatter(range(len(frequencies[column])), frequencies[column])
  mplt.xlabel('Column of {}'.format(column))
  mplt.ylabel('Frequency')
  mplt.title('Frequency Scatter Plot for {}'.format(column))
  mplt.show()
```
